```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def chi_square_metrics(expected, observed) -> float:
    #print(observed.var()), .var()
    return ((expected - observed) ** 2 / observed).sum()
```
